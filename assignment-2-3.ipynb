# Question 1:Show a breakdown of distance from home by job role and attrition

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = sns.barplot(x = df['JobRole'], y = df['DistanceFromHome'], hue = df['Attrition'])
ax.set_xticklabels(ax.get_xticklabels(), rotation = 'vertical')
ax.set_title('Distance from home by jobrole & Attrition')
```

```python
#Compare average monthly income by education and attrition.

```

```python
ax = sns.lineplot(x = df['Education'], y = df['MonthlyIncome'], hue = df['Attrition'], markers = 'True')
plt.grid()
plt.legend(loc = 'upper left')
ax.set_title('monthly income by education and attrition')
```
